```python
import pandas as pd
```

```python
# load dataset
df_sepsis = pd.read_csv('Sepsis_data2.csv')
```

1. Taking only HADM_ID withiout Sepsis

```python
df_sepsis0=df_sepsis[df_sepsis.IS_SEPSIS==0]
```

Taking only unique 20 HADMID

```python
sepsis0_hadm_ids = df_sepsis0['HADM_ID'].unique()[:20]
```

```python
sepsis0_uniq_df = df_sepsis0[df_sepsis0['HADM_ID'].isin(sepsis0_hadm_ids)]
```

Taking with Abnormal Flag

```python
#Taking only ITEMID with abnormal FLAG
sepsis0_abnormal_df=sepsis0_uniq_df[sepsis0_uniq_df['FLAG'] == 'abnormal']
```

```python
unique_columns = ['SUBJECT_ID', 'HADM_ID', 'ITEMID','LABEL']
sepsis0_abnormal = sepsis0_abnormal_df.drop_duplicates(subset=unique_columns)
```

```python
sepsis0_abnormal
```

Downloading csv file without sepsis with Abnormal Flag

```python
sepsis0_abnormal.to_csv('WithoutSepsisAbnormal.csv', index=False)
```

Without Sepsis with Normal Flag

```python
#Taking only ITEMID without abnormal FLAG
sepsis0_normal_df= sepsis0_uniq_df[sepsis0_uniq_df['FLAG'].isna()]
```

```python
unique_columns = ['SUBJECT_ID', 'HADM_ID', 'ITEMID','LABEL']
sepsis0_normal = sepsis0_normal_df.drop_duplicates(subset=unique_columns)
```

```python
sepsis0_normal
```

Downloading csv file without sepsis with normal flag

```python
sepsis0_normal.to_csv('WithoutSepsisNormal.csv', index=False)
```
